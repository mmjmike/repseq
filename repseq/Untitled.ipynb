```python
import os
all_func = []
for f in os.listdir():
    if f[-3:] == '.py':
        f = f[:-3]
        if f != '__init__':
            all_func.append(f)
for f in sorted(all_func):
    print('##', '**' + f + '**')
    print('#### ::: ' + f)
    print(f'''    options:
           show_root_toc_entry: false
    ''')
```
